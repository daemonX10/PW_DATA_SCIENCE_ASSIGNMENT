# Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed and average_of_vehicle.  
Sure, let's create a more complex example using the `Vehicle` class. We'll add some methods to this class and also create a `Car` class that inherits from `Vehicle`.

```python
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, mileage):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.mileage = mileage

    def vehicle_info(self):
        return f"Vehicle Name: {self.name_of_vehicle}, Maximum Speed: {self.max_speed}, Mileage: {self.mileage}"

    def honk(self):
        return f"{self.name_of_vehicle} is honking!"

class Car(Vehicle):
    def __init__(self, name_of_vehicle, max_speed, mileage, airbags):
        super().__init__(name_of_vehicle, max_speed, mileage)
        self.airbags = airbags

    def car_info(self):
        return f"{super().vehicle_info()}, Airbags: {self.airbags}"

# Creating an object of the Vehicle class
truck = Vehicle("Truck", 120, 8)
print(truck.vehicle_info())  # Output: Vehicle Name: Truck, Maximum Speed: 120, Mileage: 8
print(truck.honk())  # Output: Truck is honking!

# Creating an object of the Car class
car = Car("Ferrari", 220, 15, 4)
print(car.car_info())  # Output: Vehicle Name: Ferrari, Maximum Speed: 220, Mileage: 15, Airbags: 4
print(car.honk())  # Output: Ferrari is honking!
```

In this example, `Vehicle` is the parent class and `Car` is the child class. The `Car` class inherits from the `Vehicle` class and thus has access to its methods. The `Car` class also has an additional attribute `airbags`. The `super().__init__()` line in the `Car` class is used to call the `__init__` method of the `Vehicle` class, allowing us to initialize `name_of_vehicle`, `max_speed`, and `mileage` in the `Car` class.

# Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance. 
Sure, let's create a more complex example of multiple inheritance using a real-world scenario. We'll create classes for `Person`, `Employee`, `Student`, and `Intern`. An `Intern` is both an `Employee` and a `Student`.

```python
class Person:
    def __init__(self, name, age):
        """
        Initializes a Person object with the given name and age.

        Args:
            name (str): The name of the person.
            age (int): The age of the person.
        """
        self.name = name
        self.age = age

    def display(self):
        """
        Displays the name and age of the person.
        """
        print(f"Name: {self.name}, Age: {self.age}")

class Employee(Person):
    def __init__(self, name, age, company):
        """
        Initializes an Employee object with the given name, age, and company.

        Args:
            name (str): The name of the employee.
            age (int): The age of the employee.
            company (str): The company the employee works for.
        """
        super().__init__(name, age)
        self.company = company

    def display(self):
        """
        Displays the name, age, and company of the employee.
        """
        super().display()
        print(f"Company: {self.company}")

class Student(Person):
    def __init__(self, name, age, university):
        """
        Initializes a Student object with the given name, age, and university.

        Args:
            name (str): The name of the student.
            age (int): The age of the student.
            university (str): The university the student attends.
        """
        super().__init__(name, age)
        self.university = university

    def display(self):
        """
        Displays the name, age, and university of the student.
        """
        super().display()
        print(f"University: {self.university}")

class Intern(Employee, Student):
    def __init__(self, name, age, company, university):
        """
        Initializes an Intern object with the given name, age, company, and university.

        Args:
            name (str): The name of the intern.
            age (int): The age of the intern.
            company (str): The company the intern works for.
            university (str): The university the intern attends.
        """
        Employee.__init__(self, name, age, company)
        Student.__init__(self, name, age, university)

    def display(self):
        """
        Displays the name, age, company, and university of the intern.
        """
        Employee.display(self)
        print(f"University: {self.university}")

intern = Intern("John Doe", 20, "Tech Corp", "Harvard")
intern.display()
```

In this example, `Intern` is a class that inherits from both `Employee` and `Student`. It can access the `display` method from both parent classes. The `super()` function is used to call the constructor of the `Person` class from the `Employee` and `Student` classes. The `Employee.__init__()` and `Student.__init__()` are used to call the constructors of the `Employee` and `Student` classes from the `Intern` class.

# Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this class. 
In Python, getter and setter methods are used for object-oriented programming to control the access to class attributes. 

- A `getter` method is used to read the value of a private attribute. It starts with `@property` decorator, followed by a method.
- A `setter` method is used to set or update the value of an attribute. It starts with `@<attribute>.setter` decorator, followed by a method that has the same name as the attribute.



```python
class Product:
    def __init__(self, name=None, price=0, quantity=0):
        self._name = name
        self._price = price
        self._quantity = quantity

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        if not isinstance(value, str):
            raise ValueError("Name must be a string.")
        self._name = value

    @property
    def price(self):
        return self._price

    @price.setter
    def price(self, value):
        if value < 0:
            raise ValueError("Price cannot be negative.")
        self._price = value

    @property
    def quantity(self):
        return self._quantity

    @quantity.setter
    def quantity(self, value):
        if value < 0:
            raise ValueError("Quantity cannot be negative.")
        self._quantity = value

product = Product()
product.name = "Apple"  # Using setter to set the name
product.price = 1.5  # Using setter to set the price
product.quantity = 10  # Using setter to set the quantity

print(product.name)  # Using getter to get the name
print(product.price)  # Using getter to get the price
print(product.quantity)  # Using getter to get the quantity
```

In this example, `name`, `price`, and `quantity` are properties of the `Product` class. The `@property` decorator makes the methods act like getters and the `@<attribute>.setter` decorators allow us to use the methods as setters. The setter methods also include checks to ensure that the `name` is a string, and `price` and `quantity` are not negative.
